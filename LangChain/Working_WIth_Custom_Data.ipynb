```python
from  langchain_ollama.llms import OllamaLLM
```

```python
model = OllamaLLM(model="deepseek-r1:7b", temperature=0.1)
```

```python
message = [
    ('system',"you are a helpful assistant"),
    ('human',"What is the capital of France?"),
]
response = model.invoke(message)
print(response)
```
